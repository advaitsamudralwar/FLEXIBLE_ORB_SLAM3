```python
import matplotlib.pyplot as plt
import json
import seaborn as sns
import pandas as pd
# plt.rcParams["figure.figsize"] = (15,3)
buffer_data = json.load(open('BufferData.json', 'r'))
df = pd.DataFrame(buffer_data)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x="frameID", y="tracking_metric", data=df, ax=ax)
# sns.set(font_scale=1)
# sns.set(rc={'figure.figsize':(1)})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x="frameID", y="stepSize", data=df, ax=ax)
# sns.set(font_scale=2.5)
# sns.set(rc={'figure.figsize':(15,15)})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x="frameID", y="bufferLength", data=df, ax=ax)
# sns.set(font_scale=2.5)
# sns.set(rc={'figure.figsize':(30,15)})
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.lineplot(x="frameID", y="velocity", data=df, ax=ax)
# sns.set(font_scale=2.5)
# sns.set(rc={'figure.figsize':(30,15)})
plt.show()
```
