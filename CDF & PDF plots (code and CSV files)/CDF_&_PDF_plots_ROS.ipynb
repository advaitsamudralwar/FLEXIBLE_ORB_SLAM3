```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline    
```

MH01 EuRoC using ROSbag

```python
from pandas import *
 
data = read_csv("Time_ROS_MH01_Rosbag.csv")
frame = data['Frame'].tolist()
process_time = data['Processing Time'].tolist()
frame_count = len(frame)
print('Number of frames processed:', frame_count)
print('Frame ID:', frame)
print('Processing time per frame:', process_time)
```

```python
count, bins_count = np.histogram(process_time, bins=10)
# print(count)
# print(bins_count)
pdf = count / sum(count)
cdf = np.cumsum(pdf)

plt.hist(process_time, color="brown", edgecolor="white", bins=10)
plt.title("Sensor: Monocular ROS   Sequence: Euroc MH01")
plt.xlabel("Processing Time (Seconds)")
plt.ylabel("Frames")
plt.show()
```

```python
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.title("Monocular ROS on Euroc MH01")
plt.xlabel("Processing Time (Seconds)")
plt.ylabel("Probability Density")
plt.legend()
```
